A simple notebook to visualize the results of the performance parameter sweep.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('param_sweep_data.csv')
```

```python
data.iloc[0, 1:]
```

```python
for row in range(data.shape[0]):
    plt.plot(data.iloc[row, 1:].index.astype(int), data.iloc[row, 1:].values)
    plt.xlabel('Number of Attributes')
    plt.ylabel('Avg Runtime (s)')
    plt.legend(labels=data.iloc[:,0], title='Number of Rows')
```

```python
for col in range(1, data.shape[1]):
    plt.plot(data.iloc[:, 0].values.astype(int), data.iloc[:, col].values)
    plt.xlabel('Number of Rows')
    plt.ylabel('Avg Runtime (s)')
    plt.legend(labels=data.columns[1:], title='Attributes')
```
